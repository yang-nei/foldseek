```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

#  Google Drive
drive.mount('/content/drive')

#  Google Drive  Excel 
file_path = '/content/drive/My Drive/7x86re.xlsx'

#  Excel 
df = pd.read_excel(file_path)

#  x  y 
x = df['File Name']
y = df['0.8ipTM + 0.2pTM']

# 
name_mapping = {
    'SSNE5': ('SSNE5', 'pink'),
    'SSNE3': ('SSNE3', 'darkgreen'),
    'EXM34257': ('FoEG1', 'darkblue'),
    'XP_003854605.1': ('ZT9', 'yellow'),
    'XP_003857015.1': ('ZT12', 'lightgrey'),
    'XP_018152473': ('ChEC91', 'navajowhite')
}

# 
x_labels = []
colors = []
for name in x:
    found_labels = []
    found_color = 'black'  # 
    for keyword, (text, color) in name_mapping.items():
        if keyword in name:
            found_labels.append(text)  # 
            found_color = color
    if not found_labels:
        found_labels.append(name)
    x_labels.append(', '.join(found_labels))
    colors.append(found_color)

# 
plt.scatter(range(len(x)), y, c=colors)  # X

# 
plt.xlabel('')
plt.ylabel('0.8ipTM+0.2pTm')
plt.title('6Effectorsx78RLPs')

# X
plt.xticks(range(len(x)), x_labels, rotation=45, ha='center', fontsize=8)

# 
plt.gcf().set_size_inches(12, 6)

# Y
plt.ylim(0, 1)

# X
x_positions = [185, 635, 1065, 1515, 1965, 2390]  # X
plt.xticks(x_positions, [x_labels[x_pos] for x_pos in x_positions], fontsize=10)

# 
plt.savefig('/content/drive/My Drive/6x78RLP_Choice2.png')

# 
plt.tight_layout()
plt.show()
```
