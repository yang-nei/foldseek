```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

# 挂载 Google Drive
drive.mount('/content/drive')

# 更新文件路径以指向您 Google Drive 中的 Excel 文件
file_path = '/content/drive/My Drive/7x86re.xlsx'

# 读取 Excel 文件
df = pd.read_excel(file_path)

# 提取 x 和 y 变量的数据
x = df['File Name']
y = df['0.8ipTM + 0.2pTM']

# 根据文件名内容映射到所需的文字和颜色
name_mapping = {
    'SSNE5': ('SSNE5', 'pink'),
    'SSNE3': ('SSNE3', 'darkgreen'),
    'EXM34257': ('FoEG1', 'darkblue'),
    'XP_003854605.1': ('ZT9', 'yellow'),
    'XP_003857015.1': ('ZT12', 'lightgrey'),
    'XP_018152473': ('ChEC91', 'navajowhite')
}

# 将文件名中的特定关键词替换为相应的文字和颜色
x_labels = []
colors = []
for name in x:
    found_labels = []
    found_color = 'black'  # 默认为黑色
    for keyword, (text, color) in name_mapping.items():
        if keyword in name:
            found_labels.append(text)  # 只保留原始关键词
            found_color = color
    if not found_labels:
        found_labels.append(name)
    x_labels.append(', '.join(found_labels))
    colors.append(found_color)

# 绘制散点图
plt.scatter(range(len(x)), y, c=colors)  # X轴使用索引值

# 添加标签和标题
plt.xlabel('')
plt.ylabel('0.8ipTM+0.2pTm')
plt.title('6Effectorsx78RLPs')

# 使用交错显示X轴标签
plt.xticks(range(len(x)), x_labels, rotation=45, ha='center', fontsize=8)

# 增加图形的大小
plt.gcf().set_size_inches(12, 6)

# 调整Y轴范围
plt.ylim(0, 1)

# 设置X轴标签位置
x_positions = [185, 635, 1065, 1515, 1965, 2390]  # 自定义X轴标签位置
plt.xticks(x_positions, [x_labels[x_pos] for x_pos in x_positions], fontsize=10)

# 保存图形到文件
plt.savefig('/content/drive/My Drive/6x78RLP_Choice2.png')

# 显示图形
plt.tight_layout()
plt.show()
```
